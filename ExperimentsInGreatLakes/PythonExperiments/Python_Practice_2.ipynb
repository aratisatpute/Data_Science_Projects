   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm

### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
tempdf = Iris.drop('species', axis=1) # Removing class column which is not numerical

for i in tempdf.columns :
    print("Range of " + i + " is:", round(tempdf[i].max() - tempdf[i].min(),1)) 

# Rounding only to 1 decimal point as values are also rounded to 1 decimal point
```
